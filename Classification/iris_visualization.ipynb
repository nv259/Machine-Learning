```python
from sklearn.datasets import load_iris

iris = load_iris()
```

Convert iris dataset to pandas data frame

```python
import numpy as np
import pandas as pd

pd_iris = pd.DataFrame(data= np.c_[ iris['data'], iris['target'] ], columns= iris['feature_names'] + ['target'])
```

```python
pd_iris.head
```

```python
pd_iris.plot(kind='scatter', x='petal length (cm)', y='petal width (cm)')
```

### We can do something with the matplotlib library to plot in the more comprehensive way.

```python
X = iris.data[:, 2:] # petal length and width
y = iris.target
```

```python
def splitCartFromData(arr, target):
    return X[arr == target]
```

```python
import matplotlib.pyplot as plt 

XSetosa = splitCartFromData(y, 0)
XVersicolor = splitCartFromData(y, 1)
XVirginica = splitCartFromData(y, 2)

plt.scatter(XSetosa[:, 0], XSetosa[:, 1], color="red")
plt.scatter(XVersicolor[:, 0], XVersicolor[:, 1], color="yellow", marker='^')
plt.scatter(XVirginica[:, 0], XVirginica[:, 1], color="green", marker='1')

plt.legend(["Setosa", "Versicolor", "Virginica"])
plt.xlabel("Petal length")
plt.ylabel("Petal width")
```

```python
from pandas.plotting import scatter_matrix

attributes = ["petal length (cm)", "petal width (cm)" , "target"]
scatter_matrix(pd_iris[attributes], figsize = (12, 8))
```

```python
corr_matrix = pd_iris.corr()
corr_matrix["target"].sort_values(ascending=False)
```

```python
pd_iris.plot(kind='scatter', x='petal width (cm)', y='target', alpha = 0.1)
```

### *We can see that the target is much more correlated with the petal width and petal length than the others. So that, we can use these features to predict the outcome.*
