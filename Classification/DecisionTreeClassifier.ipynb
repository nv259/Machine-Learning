```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

iris = load_iris()
X = iris.data[:, 2:] # petal length and width
y = iris.target

tree_clf = DecisionTreeClassifier(max_depth=2)
tree_clf.fit(X, y)
```

```python
tree_clf_no_maxdepth = DecisionTreeClassifier()
tree_clf_no_maxdepth.fit(X, y)

tree.plot_tree(tree_clf_no_maxdepth)
```

```python
tree_clf_no_maxdepth = DecisionTreeClassifier(criterion="entropy")
tree_clf_no_maxdepth.fit(X, y)

tree.plot_tree(tree_clf_no_maxdepth)
```

```python
from sklearn import tree

tree.plot_tree(tree_clf)
```

```python
from sklearn.tree import export_graphviz
from graphviz import Source

f = open("./iris_tree.dot", "w")

export_graphviz(
    tree_clf, 
    out_file=f,
    feature_names=iris.feature_names[2:],
    class_names=iris.target_names,
    rounded=True,
    filled=True
)
```

```python
def splitCartFromData(arr, target):
    return X[arr == target]
```

```python
import matplotlib.pyplot as plt 

XSetosa = splitCartFromData(y, 0)
XVersicolor = splitCartFromData(y, 1)
XVirginica = splitCartFromData(y, 2)

plt.scatter(XSetosa[:, 0], XSetosa[:, 1], color="red")
plt.scatter(XVersicolor[:, 0], XVersicolor[:, 1], color="yellow", marker='^')
plt.scatter(XVirginica[:, 0], XVirginica[:, 1], color="green", marker='1')

plt.legend(["Setosa", "Versicolor", "Virginica"])
plt.xlabel("Petal length")
plt.ylabel("Petal width")

plt.plot([2.45, 2.45], [0, 3], ls="solid", color="red", label="Depth=0")
plt.text(1.4, 1, 'Depth=0')

plt.plot([2.45, 8], [1.75, 1.75], ls="dashed", label="Depth=1")
plt.text(2.7, 1.9, 'Depth=1')

# plt.plot([2.45, 2.45], [0, 3], ls="solid", color="red", label="Depth=0")
# plt.text(1.45, 1, 'Depth=0')

plt.show()
```

```python
def predict(features):
    val = tree_clf.predict(features)
    if val == 0:
        print("Setosa")
    if val == 1:
        print("Versicolor")
    if val == 2:
        print("Virginica")
```
